### Análisis JSON
Vamos a leer el archivo XML de la web https://www.w3schools.com/xml/cd_catalog.xml y parsearlo con la librería ElementTree

```python
#Implementamos las librerias que vamos a usar para registrar la informacion
import urllib.request
import xml.etree.ElementTree as ET

#url que analizaremos. Ya se encuentra en formato XML
url = 'https://www.w3schools.com/xml/cd_catalog.xml'

#almacenamos en "data" la informacion del link
data = urllib.request.urlopen(url).read()
tree = ET.fromstring(data.decode())

#desde el origen "DC" es la separación de cada atributo 
lst = tree.findall('CD')
print('Número de registros:', len(lst))

#generamos una lista con los valores y le añadimos titulos personales
for item in lst:
    print('\r')
    print('Titulo:', item.find('TITLE').text)
    print('Artista:', item.find('ARTIST').text)
    print('Precio:', item.find('PRICE').text)
    
#nos arroja los valores de una manera mas entendible
```
